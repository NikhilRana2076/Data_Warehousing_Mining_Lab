### Data Discretization:
Data discretization is a preprocessing technique that involves transforming continuous data into discrete intervals or categories. This process simplifies the analysis by reducing the complexity of continuous variables, making them easier to interpret and analyze.
The importances of data discretization are:
1. It helps capture non-linear relationships between variables, aids in the reduction of overfitting by minimizing noise and variability.

2. It improves the performance of machine learning models, particularly those that handle discrete or categorical data more effectively. 

3. It enhances the interpretability of the data, as discrete intervals represent meaningful ranges of values, facilitating a better understanding of relationships between variables and enabling clearer communication of analysis results to stakeholders.

4. It is essential for preparing data for certain algorithms like association rule mining, where discrete input data is required to uncover meaningful patterns and associations.
  
        
### Data Bining :

Data binning, also known as data discretization, is a process where continuous data is divided into discrete groups or bins. Each bin represents a range of values, and data points falling within that range are assigned to the corresponding bin. This technique simplifies the analysis of continuous data by organizing it into meaningful categories. 
The importance of data binning lies in its ability to make complex datasets more manageable and interpretable. By grouping similar values together, data binning helps identify patterns, trends, and relationships that may not be apparent in raw continuous data. It also facilitates the application of various statistical and machine learning techniques that require categorical or discrete input. Overall, data binning enhances the understanding and analysis of data, enabling better decision-making and insights across different domains.


### Feature Subset Selection: 
Feature subset selection, also known as feature selection, is the process of selecting a subset of relevant features from a larger set of available features in a dataset. The importance of feature subset selection lies in its ability to improve the performance, efficiency, and interpretability of machine learning models.
Certainly! Here are the key points summarizing the importance of feature subset selection:

1. Selecting relevant features can lead to improved model performance by reducing overfitting, enhancing generalization, and increasing the model's predictive accuracy.

2. Feature subset selection reduces the complexity of the model by focusing on the most informative features, making the model easier to interpret and understand.

3. Using a smaller subset of features can significantly reduce the computational resources and time required for model training and inference, making the process more efficient.

4. By removing irrelevant or redundant features, feature subset selection helps the model generalize better to unseen data, improving its ability to make accurate predictions on new observations.

5. Selecting a smaller subset of features can lead to cost savings in terms of data storage, computation, and data acquisition, particularly in large-scale or high-dimensional datasets.

Or if we want variability in the reduced feature set we select those features that are related in least way i.e. select attributes whose pairwise co-relation is as low as possible– eg Age and Dob have very high dependency on each other – don’t select both of them but DOB and Medical History may have low co-relation and you can choose this set of attributes having low co-relation out of these three ones. </b>


# Filter Approach 
### Below code shows an example of "Filter Approach" to attribute selection using CHI square test.
****
In Chi Sqaure test we see the corerelation of each atrribute with the output attribute and attributes having high correaltion with target variable are selected.Here we select the first "m" attributes that are highy corelated with the output variable. 
 

```python
dbts_new = pd.read_csv('C:/Users/acer/nikhil/DataMiningLab/lab/imputed_data_diabetes1.csv')
from sklearn.feature_selection import SelectKBest, chi2

new_dtaset = dbts_new.values
#  split the dataset into input and output variables.Since we are creating a subset of only the input or independent variables 
X = new_dtaset[:,0:8]                  # select 8 input variables. 
Y = new_dtaset[:,8]                  # select last  output variable
test = SelectKBest(score_func=chi2, k =5)   # function to get first k = 5 highest chisqaured input feature scores
fit = test.fit(X,Y)                        # Run score function on (X, Y) and get the appropriate features

# show all chisquared value/score for each input attribute
for i,j in enumerate(fit.scores_):
 print('Input Feature: %0d, Score: %.5f' % (i,j))


# Reduce X input fetaures = (9 input fetaures) to highest chisquared input fetaures K = 5 in this dataset
dbts_ftr_sbset = fit.transform(X) 

# summarize selected input features  value from the cleaned table after mean in this case
print(dbts_ftr_sbset[0:5,:])   
```

### Result Interpretation of the CHI Square score value 
From the chisquare test, we found the following attributes are less corelated to the output variable "diabetes".
****
Feature: 2 with score = 42.58251 is  "bp" 
****
Feature: 3 with score = 94.24570 is "skin"
**** 
Feature: 6 with score: 5.39268  is "Pedigree"

Hence we can reduce the 8 feature set {pregnant,glucose,bp,skin,insulin,bmi,pedigree,age} into 5 features as {pregnant,glucose,insulin,bmi,age} that's stuitable for the  learning algorithm.
