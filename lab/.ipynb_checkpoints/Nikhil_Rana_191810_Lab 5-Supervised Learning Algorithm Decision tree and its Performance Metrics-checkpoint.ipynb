### Supervised Learning: 
Supervised learning is a type of machine learning where the algorithm learns from labeled data, meaning each training example consists of input features and their corresponding target labels. The goal is to learn a mapping from input features to target labels, enabling the algorithm to make predictions on unseen data.

### Classification versus Regression Problem:
In classification problems, the goal is to predict the categorical class labels of instances based on input features. In regression problems, the goal is to predict continuous numerical values or quantities based on input features.

### Decision Tree: 
A decision tree is a supervised learning algorithm used for classification and regression tasks. It partitions the feature space into regions based on the values of input features, and at each node of the tree, a decision is made based on a feature to split the data into subsets.

### Entropy:
Entropy is a measure of uncertainty or disorder in a dataset. In the context of decision trees, entropy is used as a metric to determine the purity of a node, with lower entropy indicating higher purity or homogeneity of class labels within the node.

### Information Gain:
Information gain is a measure used in decision trees to quantify the effectiveness of a feature in reducing uncertainty or entropy. It represents the amount of information gained about the target variable by splitting the data based on a particular feature.



### $ Entropy = \sum_{i}\;-P_{i}log_{2}P_{i} $    ,   $\;\;\;\;\;\;$ $ where  P_{i} \; is \;  the \; probability\; of\; class \;i$

### Information Gain = Entropy(Parent) - [average entropy (children)]

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dbts_new = pd.read_csv('/Users/pramilakhadka/Desktop/DWDM/imputed_data_diabetes1.csv')
dbts_new.head()
```

```python
from sklearn.model_selection import train_test_split
spltd_data = dbts_new.values
# separate the dataset into input and output components
X = spltd_data [:,0:8]
Y = spltd_data[:,8]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

## Now, lets apply the cleaned data into one of the learning algorithm called Decision Tree using sklearn library.

```python
from sklearn.tree import DecisionTreeClassifier
algo = DecisionTreeClassifier(criterion = "entropy",random_state = 42, max_depth = 4 , min_samples_leaf=5)  # decision tree witha depth of 4
algo.fit(X_train, Y_train)
result = algo.score(X_test, Y_test)
print(("The Decision Tree model has given Accuracy of: %.3f%%") % (result*100.0))
```

### Decision Tree Plot

```python
from sklearn import tree
plt.figure(figsize=(30,30))
tree.plot_tree(algo) 
```

## Performance Metrics of learning algorithm

### Evaluate the predictive accuracy of a model


### Create Confusion Matrix to compare the number of predicted values against actual labeled value 

```python
Y_Prdct = algo.predict(X_test)
print(Y_Prdct)
```

```python
from sklearn import metrics
cf_mtrx = metrics.confusion_matrix(Y_test, Y_Prdct)
cf_mtrx
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cf_mtrx, display_labels = [ False,True]) #[False, True])
cm_display.plot()
plt.show() 
```
